## **Preparaci贸n de los datos para los algoritmos de Machine Learning**

En este apartado vamos a limpiar el dataset para dejarlo listo para los modelos que elijamos.

Vamos a indicar cu谩les son las stopwords y a habilitar el lematizador.

```python
stop_words = set(stopwords.words("english"))
lemmatizer = WordNetLemmatizer()
```

Definimos las etiquetas para los tokens.

```python
def get_wordnet_pos(tag):
    if tag.startswith("J"):
        return wordnet.ADJ
    elif tag.startswith("V"):
        return wordnet.VERB
    elif tag.startswith("N"):
        return wordnet.NOUN
    elif tag.startswith("R"):
        return wordnet.ADV
    return wordnet.NOUN
```

Funci贸n para limpiar y lematizar el texto.

```python
def clean_text(text):
    text = text.lower()
    text = re.sub(r"http[s]?://\S+|www\.", " ", text) # Regex para quitar urls
    text = re.sub(r"\d+", " ", text)            # Quitar n煤meros
    text = re.sub(r"[^a-zA-Z\s]", " ", text)    # Quitar simbolos y puntuaciones
    text = re.sub(r"\s+", " ", text).strip()    # Quitar espacios extra a帽adidos
    #  quitar la \n y \r y alguna paranoia hacia el texto por el estilo  bombarden la rae
    return tokenize(text)
```

```python
def tokenize(text):
    tokens = word_tokenize(text)
    tags = pos_tag(tokens)

    return " ".join([lemmatizer.lemmatize(word, get_wordnet_pos(tag))
                        for word, tag in tags if word not in stop_words])
```

Aplicamos la limpieza y tokenizaci贸n de los textos.

```python
emails_df["text"] = emails_df["text"].apply(clean_text)
```

Preparamos el **vectorizado**, al final vamos a usar el `TfidfVectorizer` al cual vamos a personalizarlo usando distintos parametros.

```python
tfidf = TfidfVectorizer(
    ngram_range=(1, 2),     # Aceptar unigramas y bigramas como "free" y "free money"
    max_df = 0.9,           # Ratio m谩ximo de aparici贸n para ser contabilizado en el diccionario
    min_df = 2,             # Ratio m铆nimo de aparici贸n para ser a帽adida al diccionario
    max_features = 5000,    # Tama帽o m谩ximo del diccionario
    dtype = np.float32,     # Formato
    sublinear_tf = True     # Cambia la forma de contar las aparaciones para evitar distorsiones pasa de lineal a logar铆tmica
)
```

### **Particiones para entrenamiento y prueba**

En este subapartado se dividir谩 el dataset en $X$ y $y$, y posteriormente en sus dos subconjuntos de entrenamiento y prueba.

Y en sus dos sub divisiones para entrenamiento y pruebas.

* $X$: `text` tokenizado.

* $y$: Nuestro target `is_spam`.

```python
X = tfidf.fit_transform(emails_df["text"])
y = emails_df["is_spam"]
```

Separamos el dataset en dos partes:

* Datos de entranamiento (training set): 80%
* Datos de prueba (test set): 20% (si tenemos muchos datos, el porcentaje seria menor)

Adem谩s, utilizaremos `random_state = 777` para que en todos los entrenamientos y momentos los conjuntos de entrenamiento y prueba sean siempre los mismos.

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size = 0.2, random_state = 777
)
```

## **Exposici贸n del objetivo**

Nuestro objetivo es averiguar si un email es spam o no, mediante nuestro target que sera la **`is_spam`** mediante un clasificador.

## **Entrenamiento del modelo y comprobaci贸n del rendimiento**

En este apartado se entrenar谩n varios modelos, comprobando en cada caso el rendimiento que ofrecen mediante las apropiadas medidas de error y/o acierto.

Vamos a probar varios modelos supervisados de clasificaci贸n para determinar cu谩l nos da el mejor resultado.

* **MultinomialNB**
* **BernoulliNB**
* **LogisticRegression**
* **LinearSVC**
* **RandomForestClassifier**
* **KNeighborsClassifier**


```python
model_nb_multinominal = MultinomialNB()
model_nb_bernoulli = BernoulliNB()
model_lr_classifier = LogisticRegression(max_iter=1000)
model_svc = LinearSVC()
model_rf_classifier = RandomForestClassifier(n_estimators=100)
model_knn = KNeighborsClassifier(n_neighbors=5)
```

Para evaluar los modelos, utilizaremos la siguiente funci贸n.

```python
def test_models(models_dict):
    rows, cols = 2, 3

    plt.figure(figsize=(15, 10))

    for i, (name, y_pred) in enumerate(models_dict.items(), 1):
        print(name)
        print(f"Accuracy: {accuracy_score(y_test, y_pred):.4f}\n")
        print(classification_report(y_test, y_pred, digits=4))

        # Confusion matrix
        cm = confusion_matrix(y_test, y_pred)

        ax = plt.subplot(rows, cols, i)
        sns.heatmap(
            cm,
            annot=True,
            fmt="d",
            cmap="Blues",
            xticklabels=["ham", "spam"],
            yticklabels=["ham", "spam"],
            ax=ax
        )
        ax.set_title(name)
        ax.set_xlabel("Predicci贸n")
        ax.set_ylabel("Etiqueta real")

    plt.tight_layout()
    plt.show()
```

Entrenamos los modelos y guardamos sus resultados.

```python
model_nb_multinominal.fit(X_train, y_train)
y_nb_multinominal = model_nb_multinominal.predict(X_test)
```

```python
model_nb_bernoulli.fit(X_train, y_train)
y_nb_bernoulli = model_nb_bernoulli.predict(X_test)
```

```python
model_lr_classifier.fit(X_train, y_train)
y_lr_classifier = model_lr_classifier.predict(X_test)
```

```python
model_svc.fit(X_train, y_train)
y_svc = model_svc.predict(X_test)
```

```python
model_rf_classifier.fit(X_train, y_train)
y_rf_classifier = model_rf_classifier.predict(X_test)
```

```python
model_knn.fit(X_train, y_train)
y_knn = model_knn.predict(X_test)
```

### **Probar modelos:**

```python
test_models({
    "MultinomialNB": y_nb_multinominal,
    "BernoulliNB": y_nb_bernoulli,
    "LogisticRegression": y_lr_classifier,
    "LinearSVC": y_svc,
    "RandomForestClassifier": y_rf_classifier,
    "KNeighborsClassifier": y_knn
})
```

## **Conclusiones**

Despu茅s de evaluar todos los modelos, el que tuvo mejores resultados fue el **`LinearSVC`** con una precisi贸n del **98.65%**, tambi茅n muy cerca tenemos al **`Logistic Regression`**, ambos tienen mejores resultado que el resto de algoritmos.

Podemos sacar las siguientes conclusiones:

* En general, todos los modelos, excepto el **`KNN`**, muestran un buen rendimiento, teniendo resultados altos en `accuracy`, `precision`, `recall` y `f1-score`.
* El `accuracy`, que muestra la exactitud de nuestro modelo midiendo el porcentaje total de predicciones correctas, los mejores fueron el **`LinearSVC`** y el **`Logistic Regression`**, seguidos por el **`Random Forest`**. Podemos ver que el peor en este apartado es el **`KNN`** con un **57%** de acierto, lo cual muestra una bajada de exactitud notable comparado con los dem谩s, que no bajan del **90%**.
* La `precision`, indica cu谩ntas predicciones positivas fueron correctas (evita falsos positivos), los modelos lineales son los mejores en este apartado. **`KNN`** y **`Naive Bayes`** sufren al detectar el spam.
* El `recall`, mide cu谩ntos casos reales detecta correctamente (evita falsos negativos), aqu铆 todos muestran valores buenos excepto el **`KNN`**, que predice casi todo como spam.
* El `f1-score`, que combina `precision` y `recall` en una sola m茅trica, el algoritmo `LinearSVC` es el m谩s gen茅rico siendo el mejor en este categor铆a.
* El `support`, que indica cu谩ntas muestras hay por clase, tenemos:
    * Clase 0 (no-spam): 732
    * Clase 1 (spam): 303

* En `macro avg` y `weighted avg`, `macro avg` cuenta todas las clases por igual y `weighted avg` las valora seg煤n el n煤mero de muestras, **`LinearSVC`** y **`Logistic Regression`** tienen valores casi id茅nticos, se帽al de estabilidad. **`Random Forest`** y **`Naive Bayes`** tienen buenos valores, pero **`KNN`** destaca por ser el peor aqu铆 tambi茅n.
* No parece que haya overfitting, aunque existe una posibilidad leve en el **`Random Forest`** debido a su complejidad y que no est谩 limitado.

Con lo dicho nos quedaremos con el modelo **`LinearSVC`**, por sus buenos resultado y estabilidad.
